```python

numerical_features = data.select_dtypes(include=np.number).columns

for col in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[col], kde=True)  # KDE adds a density curve to the histogram
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    # Calculate and print skewness
    skewness = data[col].skew()
    print(f'Skewness of {col}: {skewness}')
```

```python
# missing value handling

# Fill missing values in 'Gender' with the mode
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)

# Fill missing values in 'Married' with the mode
data['Married'].fillna(data['Married'].mode()[0], inplace=True)

# Fill missing values in 'Dependents' with the mode
data['Dependents'].fillna(data['Dependents'].mode()[0], inplace=True)

# Fill missing values in 'Self_Employed' with the mode
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)

# Fill missing values in 'LoanAmount' with the median
data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)

# Fill missing values in 'Loan_Amount_Term' with the mode
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0], inplace=True)

# Fill missing values in 'Credit_History' with the mode
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)

# Check for remaining missing values
data.isnull().sum()
```

```python

# Loop through each column in the DataFrame
for column in data.columns:
    # Check if the column's data type is numeric
    if pd.api.types.is_numeric_dtype(data[column]):
        # Create a histogram for the column
        plt.figure(figsize=(8, 6))
        plt.hist(data[column], bins=10, edgecolor='black')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```

**Outlier detection**

```python
# prompt: box plot for all columns using for

for column in data.columns:
  if pd.api.types.is_numeric_dtype(data[column]):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

```python

for column in data.columns:
  if pd.api.types.is_numeric_dtype(data[column]):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    if not outliers.empty:
      print(f"Outliers in column '{column}':")
      print(outliers[[column]]) # Print only the outlier values for the current column
    else:
      print(f"No outliers detected in column '{column}'")
```

```python


def clip_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values outside the bounds
    df[column] = np.clip(df[column], lower_bound, upper_bound)

for column in data.columns:
    if pd.api.types.is_numeric_dtype(data[column]):
        print(f"Processing column: {column}")
        clip_outliers(data, column)
        # Print details after clipping
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        print(f"  Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
        print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
        print(f"  Min: {data[column].min()}, Max: {data[column].max()}")
```

```python
# box plot before and after outlier clipping

import matplotlib.pyplot as plt
import seaborn as sns

for column in data.columns:
    if pd.api.types.is_numeric_dtype(data[column]):
        plt.figure(figsize=(12, 6))  # Adjust figure size for better visualization

        # Before clipping
        plt.subplot(1, 2, 1)  # Create subplots for side-by-side comparison
        sns.boxplot(x=data[column], color='skyblue')  # Use a color for better visibility
        plt.title(f'Box Plot of {column} (Before Clipping)')

        # After clipping (assuming 'clip_outliers' function modifies the data in place)
        clip_outliers(data, column) #Calling the clipping function again, as it was called within a loop earlier and might not have clipped properly.
        plt.subplot(1, 2, 2)
        sns.boxplot(x=data[column], color='lightgreen')
        plt.title(f'Box Plot of {column} (After Clipping)')

        plt.tight_layout()  # Adjust subplot params for a tight layout
        plt.show()
```

**Encoding**

```python
# encode the data

# Encoding categorical features
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']:
    data[col] = le.fit_transform(data[col])

data.head()
```

**Scale**

```python
#  Scale the data


from sklearn.preprocessing import MinMaxScaler

# Select numerical columns for scaling
numerical_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the selected numerical columns
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Now the numerical features in 'data' are scaled between 0 and 1
print(data.head())
```
