##5)Perform meaningful visualizations using the dataset ?

```python

import matplotlib.pyplot as plt

# list of colors for the histograms
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightblue', 'lightpink', 'khaki', 'plum']

for i, col in enumerate(data.columns):
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    plt.hist(data[col], bins=10, color=colors[i % len(colors)], edgecolor='black') # Use colors from the list, cycling through them if there are more columns than colors
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {col}')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of Outcome (Diabetic or Not)
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=data)
plt.title('Distribution of Diabetic Patients')
plt.show()


# 2. Relationship between Glucose and Outcome
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title('Glucose Levels in Diabetic and Non-Diabetic Patients')
plt.show()


```

```python

# Iterate through all columns except 'Outcome'
for column in data.columns[:-1]:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Outcome', y=column, data=data)
    plt.title(f'Distribution of {column} vs. Outcome')
    plt.show()
```

```python


import matplotlib.pyplot as plt

# Calculate the mean of each column for each outcome group
means_by_outcome = data.groupby('Outcome').mean()

# Transpose the DataFrame for easier plotting
means_by_outcome = means_by_outcome.T

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(means_by_outcome[0], labels=means_by_outcome.index, autopct='%1.1f%%', startangle=90)
plt.title('Mean of Each Column for Outcome = 0')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

plt.figure(figsize=(10, 8))
plt.pie(means_by_outcome[1], labels=means_by_outcome.index, autopct='%1.1f%%', startangle=90)
plt.title('Mean of Each Column for Outcome = 1')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# 3. Correlation Matrix Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Diabetes Dataset Features')
plt.show()
```
